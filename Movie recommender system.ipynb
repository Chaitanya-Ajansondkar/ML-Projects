```python
import ast
# ast.literal_eval() --converts string into just list
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
new_movies['genres']=new_movies['genres'].apply(convert) #applying above function on column genres
```

```python
new_movies.head()
```

```python
new_movies['keywords']=new_movies["keywords"].apply(convert)
```

```python
new_movies.head(2)
```

```python
new_movies['cast'][0]  
```

```python
#for column cast, we select only first 3 actor names as they are important
def convert2(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L  
```

```python
new_movies['cast']=new_movies['cast'].apply(convert2)
```

```python
#for crew we need only director
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L  
```

```python
new_movies['crew']=new_movies["crew"].apply(fetch_director)
```

```python
#converting string to list
new_movies['overview']=new_movies['overview'].apply(lambda x:x.split())
```

```python
new_movies.head(3)
```

```python
#removing spaces to avoid confusion between same words
new_movies['genres']=new_movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['keywords']=new_movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['cast']=new_movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['crew']=new_movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['genres']+new_movies['overview']+new_movies['keywords']+new_movies['cast']+new_movies['crew']
```

```python
new_movies.head(3)
```

```python
final_new_movies=new_movies[['movie_id','title','tags']]
```

```python
final_new_movies.head(3)
```

```python
#convert tags list into string
final_new_movies['tags']=final_new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
final_new_movies.head(3)
```

```python
#converting to lower case
final_new_movies['tags']=final_new_movies['tags'].apply(lambda x: x.lower())
```

```python
final_new_movies.head(3)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def new (text):
    y=[]
    for i in text.split():  # looping over string converted into list
        y.append(ps.stem(i)) #using stem function
    z=" ".join(y) #converting back to string
    return z
```

```python
final_new_movies['tags']=final_new_movies['tags'].apply(new)
```

```python
#using count vectorizer to convert text into words count
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_new_movies['tags'])
vectors
```

```python
print(vectors)
```

```python
#converting to array
vectors = vectors.toarray()
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
# get all unique words in the corpus
vocab = cv.get_feature_names()
# show document feature vectors
df = pd.DataFrame(vectors, columns=vocab)
df
```

```python
# Compute Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity=(cosine_similarity(vectors))   #vectors and df are one and the same , one is array other is dataframe
```

```python
similarity
```

```python
def recommend(movie):
    movie_index=final_new_movies[final_new_movies['title']==movie].index[0] #fetching index
    distance = similarity[movie_index]     #similarity[movie_index] gives similarity array for that particular movie index 
    dis_list= list(enumerate(distance)) # gives numbering to 'distance' array created above, so we get index no. of all movies
    distance_3= sorted(dis_list,reverse=True,key=lambda x:x[1])#there are 2 elements in distance 2 and we need order based on 2nd element
    distance_4=distance_3[1:6] #getting top 5 movie indices with highest similarity values
    for i in distance_4:
        print(final_new_movies.iloc[i[0]].title) # as distance_4 is list having index and similarity score of 5 movies
    
```

```python
recommend('Avatar') #checking the model 
```
